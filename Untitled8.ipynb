```python
x=df.iloc[:,[2,3]].values
print(x)
# age and salary
```

```python
y=df.iloc[:,4].values
print(y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=47)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier.fit(x_train,y_train)
```

```python
y_pred=classifier.predict(x_test)
print(y)
print(y_pred)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
cm1=confusion_matrix(y_test,y_pred)
print("Confusion matrix:")
print(cm1)
ac1=accuracy_score(y_test,y_pred)*100
print("Accuracy Score:")
print(ac1)
```

```python
truePositive=cm1[0][0]
trueNegative=cm1[1][1]
falsePositive=cm1[0][1]
falseNegative=cm1[1][0]
total=truePositive+trueNegative+falsePositive+falseNegative
```

```python
error_rate=(falsePositive+falseNegative)/total
print("error rate:")
print(error_rate)
```

```python
from sklearn.metrics import classification_report
print("Classification Report:")
print(classification_report(y_test,y_pred))
```
